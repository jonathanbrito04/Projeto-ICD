```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('video_games.csv')


df.head()
```

```python
df
```

```python
#Limpando o dataset de linhas que possuiam o valor "NaN"
#Possuia em torno de 9.894 linhas com esse valor 
df_limpo = df.dropna()
print(df_limpo)
```

```python
# Ordenar o DataFrame df_limpo pelo número de vendas globais em ordem decrescente
df_sorted = df_limpo.sort_values(by='Vendas_Globais', ascending=False)

# Selecionar os top 100 jogos com maior venda global
top_100_jogos = df_sorted.head(100)

# Mostrar os 100 jogos com maior venda global
print(top_100_jogos)
```

```python
top_100_jogos
```

```python
# Removendo as colunas de números de votantes da nota da crítica, e a de usuário, não achavamos que seria necessário trabalhar com essas tabelas
top_100_jogos = top_100_jogos.drop(columns=['Contagem_Crítica', 'Contagem_Usuário'], errors='ignore')

# Verificar as colunas após a remoção
print("Colunas após remoção:", top_100_jogos.columns)
```

```python
top_100_jogos
```

```python

colunas_numericas = ['Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros', 'Vendas_Globais']


medidas = {}


for column in colunas_numericas:
    medidas[column] = {
        'Média': top_100_jogos[column].mean(),
        'Mediana': top_100_jogos[column].median(),
        'Desvio Padrão': top_100_jogos[column].std(),
        'Variância': top_100_jogos[column].var(),
        'Mínimo': top_100_jogos[column].min(),
        'Máximo': top_100_jogos[column].max(),
        'Primeiro Quartil (25%)': top_100_jogos[column].quantile(0.25),
        'Terceiro Quartil (75%)': top_100_jogos[column].quantile(0.75),
    }


medidas_df = pd.DataFrame(medidas).T
print(medidas_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


numerical_columns = ['Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros', 'Vendas_Globais']


plt.figure(figsize=(14, 8))


for i, column in enumerate(numerical_columns, start=1):
    plt.subplot(2, 3, i)  # Ajuste o layout conforme necessário
    sns.boxplot(x=top_100_jogos[column])
    plt.title(column)
    plt.xlabel('Valor')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
